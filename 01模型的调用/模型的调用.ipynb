```python
# 非对话模型

from langchain.chat_models import init_chat_model
import os
from dotenv import load_dotenv
load_dotenv()
# 设置环境变量（可选，init_chat_model 会用到）
api_key = os.getenv("OPENAI_API_KEY")
base_url = os.getenv("OPENAI_BASE_URL")
# 使用 init_chat_model
llm = init_chat_model(
    "openai:google/gemma-3n-e4b",  # 模型 ID
    api_key=api_key,
    base_url=base_url,
    temperature = 0.1,
)
response= llm.invoke("写一首冬天的诗")
print(response.content)
```

系统环境夹在系统变量

```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
load_dotenv()
# 环境变量自动获取了
llm = init_chat_model(
    "openai:google/gemma-3n-e4b",  # LM Studio 模型 ID
    temperature = 0.1,
    max_tokens=100, # token限制
)
response= llm.invoke("唐诗三百首")
print(response.content)
```
